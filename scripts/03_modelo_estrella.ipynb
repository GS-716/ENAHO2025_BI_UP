### 3. Función para crear dimensiones

```python
def crear_dimension(df, columnas, nombre_key):
    """
    Crea una dimensión con combinaciones únicas y surrogate key.
    """
    dim = df[columnas].copy()

    for col in columnas:
        if pd.api.types.is_numeric_dtype(dim[col]):
            dim[col] = dim[col].fillna(-1)
        else:
            dim[col] = dim[col].astype("string").fillna("No especificado")

    dim = dim.drop_duplicates().reset_index(drop=True)
    dim.insert(0, nombre_key, range(1, len(dim) + 1))

    return dim
```

### 3.1 Variables auxiliares

```python
stg_model = stg.copy()

# Rango de habitaciones totales
stg_model["rango_habitaciones_total"] = pd.cut(
    stg_model["habitaciones_total"],
    bins=[0, 1, 2, 3, 5, float("inf")],
    labels=[
        "1 habitación",
        "2 habitaciones",
        "3 habitaciones",
        "4 a 5 habitaciones",
        "6 a más habitaciones"
    ],
    include_lowest=True
).astype("string")

# Rango de habitaciones para dormir
stg_model["rango_habitaciones_dormir"] = pd.cut(
    stg_model["habitaciones_dormir"],
    bins=[0, 1, 2, 3, float("inf")],
    labels=[
        "1 dormitorio",
        "2 dormitorios",
        "3 dormitorios",
        "4 a más dormitorios"
    ],
    include_lowest=True
).astype("string")

# Categoría de carencias
stg_model["categoria_carencias"] = pd.cut(
    stg_model["total_carencias_bienestar_habitacional"],
    bins=[-1, 0, 1, 2, 4, 7],
    labels=[
        "Sin carencias",
        "Una carencia",
        "Dos carencias",
        "Tres a cuatro carencias",
        "Cinco a más carencias"
    ]
).astype("string")
```

### 8. Función para añadir claves al fact

```python
def agregar_key_fact(fact_base, stg_base, dim, columnas_dim, nombre_key):
    """
    Agrega la surrogate key de una dimensión a la fact table.
    """
    left = stg_base[columnas_dim].copy()
    right = dim.copy()

    for col in columnas_dim:
        if pd.api.types.is_numeric_dtype(left[col]):
            left[col] = left[col].fillna(-1)
            right[col] = right[col].fillna(-1)
        else:
            left[col] = left[col].astype("string").fillna("No especificado")
            right[col] = right[col].astype("string").fillna("No especificado")

    left["_row_id"] = stg_base.index

    merged = left.merge(
        right[[nombre_key] + columnas_dim],
        on=columnas_dim,
        how="left"
    )

    merged = merged.set_index("_row_id").sort_index()

    fact_base[nombre_key] = merged[nombre_key].values

    return fact_base
```

### 9. Crear base inicial del fact

```python
fact_hogar_bienestar = pd.DataFrame(index=stg_model.index)

fact_hogar_bienestar["hogar_id"] = stg_model["hogar_id"]
fact_hogar_bienestar["conglome"] = stg_model["conglome"]
fact_hogar_bienestar["vivienda"] = stg_model["vivienda"]
fact_hogar_bienestar["hogar"] = stg_model["hogar"]

fact_hogar_bienestar.head()
```

### 10. Agregar claves de dimensión al fact

```python
fact_hogar_bienestar = agregar_key_fact(
    fact_hogar_bienestar,
    stg_model,
    dim_tiempo,
    cols_dim_tiempo,
    "tiempo_key"
)

fact_hogar_bienestar = agregar_key_fact(
    fact_hogar_bienestar,
    stg_model,
    dim_geografia,
    cols_dim_geografia,
    "geografia_key"
)

fact_hogar_bienestar = agregar_key_fact(
    fact_hogar_bienestar,
    stg_model,
    dim_vivienda,
    cols_dim_vivienda,
    "vivienda_key"
)

fact_hogar_bienestar = agregar_key_fact(
    fact_hogar_bienestar,
    stg_model,
    dim_materiales_vivienda,
    cols_dim_materiales_vivienda,
    "materiales_vivienda_key"
)

fact_hogar_bienestar = agregar_key_fact(
    fact_hogar_bienestar,
    stg_model,
    dim_habitabilidad,
    cols_dim_habitabilidad,
    "habitabilidad_key"
)

fact_hogar_bienestar = agregar_key_fact(
    fact_hogar_bienestar,
    stg_model,
    dim_tenencia_propiedad,
    cols_dim_tenencia_propiedad,
    "tenencia_propiedad_key"
)

fact_hogar_bienestar = agregar_key_fact(
    fact_hogar_bienestar,
    stg_model,
    dim_agua,
    cols_dim_agua,
    "agua_key"
)

fact_hogar_bienestar = agregar_key_fact(
    fact_hogar_bienestar,
    stg_model,
    dim_saneamiento,
    cols_dim_saneamiento,
    "saneamiento_key"
)

fact_hogar_bienestar = agregar_key_fact(
    fact_hogar_bienestar,
    stg_model,
    dim_energia,
    cols_dim_energia,
    "energia_key"
)

fact_hogar_bienestar = agregar_key_fact(
    fact_hogar_bienestar,
    stg_model,
    dim_conectividad,
    cols_dim_conectividad,
    "conectividad_key"
)

fact_hogar_bienestar.head()
```

```python
cols_metricas_fact = [
    # Factor muestral
    "factor_expansion",

    # Medidas de vivienda
    "habitaciones_total",
    "habitaciones_dormir",
    "alquiler_mensual_reportado",
    "alquiler_estimado_mensual",
    "alquiler_anual_imputado",
    "alquiler_estimado_anual_imputado",

    # Agua
    "horas_agua_dia",
    "dias_agua_semana",

    # Indicadores para agregación
    "agua_red_publica_ind",
    "agua_potable_ind",
    "cloro_seguro_ind",
    "agua_todos_dias_ind",
    "agua_segura_ind",
    "saneamiento_adecuado_ind",

    "pared_precaria_ind",
    "piso_precario_ind",
    "techo_precario_ind",
    "vivienda_material_precario_ind",

    "nbi_vivienda_inadecuada_ind",
    "nbi_hacinamiento_ind",
    "nbi_sin_servicio_higienico_ind",
    "nbi_ninios_no_asisten_ind",
    "nbi_alta_dependencia_ind",

    "tiene_electricidad_ind",
    "combustible_limpio_ind",
    "combustible_solido_ind",

    "tiene_telefono_fijo_ind",
    "tiene_celular_ind",
    "tiene_tv_cable_ind",
    "tiene_internet_ind",
    "sin_tic_ind",
    "tiene_tdt_ind",

    # Gastos
    "gasto_anual_servicios_pagado_hogar",
    "gasto_anual_servicios_donado",
    "gasto_anual_servicios_autoconsumo",
    "gasto_anual_servicios_total",
    "gasto_mensual_servicios_estimado",

    # Carencias
    "carencia_materiales_vivienda_ind",
    "carencia_hacinamiento_ind",
    "carencia_agua_segura_ind",
    "carencia_saneamiento_ind",
    "carencia_electricidad_ind",
    "carencia_combustible_limpio_ind",
    "carencia_internet_ind",
    "total_carencias_bienestar_habitacional",
    "brecha_multidimensional_ind"
]

cols_metricas_fact = [col for col in cols_metricas_fact if col in stg_model.columns]

fact_hogar_bienestar[cols_metricas_fact] = stg_model[cols_metricas_fact]

fact_hogar_bienestar.head()
```
