### 1. Cargar Staging

```python
from pathlib import Path
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", 120)
pd.set_option("display.max_rows", 120)

PATH_STG = Path.cwd().parent / "data" / "processed" / "stg_enaho_hogar_100_limpio.parquet"

stg = pd.read_parquet(PATH_STG)

print("Filas staging:", len(stg))
print("Columnas staging:", len(stg.columns))

stg.head()
```

### 2. Validaciones iniciales

```python
assert "hogar_id" in stg.columns, "No existe hogar_id"
assert stg["hogar_id"].notna().all(), "Hay hogar_id nulos"
assert stg["hogar_id"].duplicated().sum() == 0, "Hay hogar_id duplicados"

print("Staging validado correctamente.")
print("Hogares:", len(stg))
```

### 3. Función para crear dimensiones

```python
def crear_dimension(df, columnas, nombre_key):
    """
    Crea una dimensión con combinaciones únicas de columnas
    y una surrogate key incremental.
    """
    dim = df[columnas].copy()

    # Para evitar problemas en merges con nulos
    for col in columnas:
        if pd.api.types.is_numeric_dtype(dim[col]):
            dim[col] = dim[col].fillna(-1)
        else:
            dim[col] = dim[col].astype("string").fillna("No especificado")

    dim = dim.drop_duplicates().reset_index(drop=True)
    dim.insert(0, nombre_key, range(1, len(dim) + 1))

    return dim
```

### 4. Crear dim_tiempo

```python
cols_dim_tiempo = [
    "anio",
    "mes",
    "periodo"
]

dim_tiempo = crear_dimension(
    stg,
    columnas=cols_dim_tiempo,
    nombre_key="tiempo_key"
)

dim_tiempo.head()
```

### 5. dim_ubicacion

```python
cols_dim_ubicacion = [
    "ubigeo",
    "dominio_cod",
    "dominio",
    "estrato_cod",
    "estrato",
    "latitud",
    "longitud",
    "altitud"
]

dim_ubicacion = crear_dimension(
    stg,
    columnas=cols_dim_ubicacion,
    nombre_key="ubicacion_key"
)

dim_ubicacion.head()
```

### 4. dim_vivienda

```python
cols_dim_vivienda = [
    "tipo_vivienda_cod",
    "tipo_vivienda",
    "material_pared_cod",
    "material_pared",
    "material_piso_cod",
    "material_piso",
    "material_techo_cod",
    "material_techo",
    "tenencia_vivienda_cod",
    "tenencia_vivienda"
]

dim_vivienda = crear_dimension(
    stg,
    columnas=cols_dim_vivienda,
    nombre_key="vivienda_key"
)

dim_vivienda.head()
```

### 5. Dim servicios básicos

```python
cols_dim_servicios = [
    "fuente_agua_cod",
    "fuente_agua",
    "nivel_cloro_cod",
    "nivel_cloro",
    "servicio_higienico_cod",
    "servicio_higienico"
]

dim_servicios_basicos = crear_dimension(
    stg,
    columnas=cols_dim_servicios,
    nombre_key="servicios_basicos_key"
)

dim_servicios_basicos.head()
```

### 6. Dim energia

```python
cols_dim_energia = [
    "combustible_cocina_cod",
    "combustible_cocina"
]

dim_energia = crear_dimension(
    stg,
    columnas=cols_dim_energia,
    nombre_key="energia_key"
)

dim_energia.head()
```

### 7. Dim conectividad

```python
cols_dim_conectividad = [
    "tiene_telefono_fijo_ind",
    "tiene_celular_ind",
    "tiene_tv_cable_ind",
    "tiene_internet_ind",
    "sin_tic_ind",
    "tiene_tdt_ind"
]

dim_conectividad = crear_dimension(
    stg,
    columnas=cols_dim_conectividad,
    nombre_key="conectividad_key"
)

dim_conectividad.head()
```

### 8. Función para añadir claves al fact

```python
def agregar_key_fact(fact_base, stg_base, dim, columnas_dim, nombre_key):
    """
    Agrega la surrogate key de una dimensión a la tabla fact.
    """
    left = stg_base[columnas_dim].copy()
    right = dim.copy()

    for col in columnas_dim:
        if pd.api.types.is_numeric_dtype(left[col]):
            left[col] = left[col].fillna(-1)
            right[col] = right[col].fillna(-1)
        else:
            left[col] = left[col].astype("string").fillna("No especificado")
            right[col] = right[col].astype("string").fillna("No especificado")

    left["_row_id"] = stg_base.index

    merged = left.merge(
        right[[nombre_key] + columnas_dim],
        on=columnas_dim,
        how="left"
    )

    merged = merged.set_index("_row_id").sort_index()

    fact_base[nombre_key] = merged[nombre_key].values

    return fact_base
```

### 9. Crear base inicial del fact

```python
fact_bienestar_habitacional = pd.DataFrame(index=stg.index)

# Llave degenerate del hogar
fact_bienestar_habitacional["hogar_id"] = stg["hogar_id"]

# Llaves naturales útiles para trazabilidad
fact_bienestar_habitacional["conglome"] = stg["conglome"]
fact_bienestar_habitacional["vivienda"] = stg["vivienda"]
fact_bienestar_habitacional["hogar"] = stg["hogar"]

fact_bienestar_habitacional.head()
```

### 10. Agregar claves de dimensión al fact

```python
fact_bienestar_habitacional = agregar_key_fact(
    fact_bienestar_habitacional,
    stg,
    dim_tiempo,
    cols_dim_tiempo,
    "tiempo_key"
)

fact_bienestar_habitacional = agregar_key_fact(
    fact_bienestar_habitacional,
    stg,
    dim_ubicacion,
    cols_dim_ubicacion,
    "ubicacion_key"
)

fact_bienestar_habitacional = agregar_key_fact(
    fact_bienestar_habitacional,
    stg,
    dim_vivienda,
    cols_dim_vivienda,
    "vivienda_key"
)

fact_bienestar_habitacional = agregar_key_fact(
    fact_bienestar_habitacional,
    stg,
    dim_servicios_basicos,
    cols_dim_servicios,
    "servicios_basicos_key"
)

fact_bienestar_habitacional = agregar_key_fact(
    fact_bienestar_habitacional,
    stg,
    dim_energia,
    cols_dim_energia,
    "energia_key"
)

fact_bienestar_habitacional = agregar_key_fact(
    fact_bienestar_habitacional,
    stg,
    dim_conectividad,
    cols_dim_conectividad,
    "conectividad_key"
)

fact_bienestar_habitacional.head()
```

### 11. Agregar métricas e indicadores al fact

```python
cols_metricas_fact = [
    # Factor muestral
    "factor_expansion",

    # Vivienda y habitabilidad
    "habitaciones_total",
    "habitaciones_dormir",
    "alquiler_mensual_reportado",
    "alquiler_estimado_mensual",
    "alquiler_anual_imputado",
    "alquiler_estimado_anual_imputado",

    # Agua
    "agua_red_publica_ind",
    "agua_potable_ind",
    "cloro_seguro_ind",
    "agua_todos_dias_ind",
    "agua_segura_ind",
    "horas_agua_dia",
    "dias_agua_semana",

    # Saneamiento
    "saneamiento_adecuado_ind",

    # Materiales y NBI
    "pared_precaria_ind",
    "piso_precario_ind",
    "techo_precario_ind",
    "vivienda_material_precario_ind",
    "nbi_vivienda_inadecuada_ind",
    "nbi_hacinamiento_ind",
    "nbi_sin_servicio_higienico_ind",
    "nbi_ninios_no_asisten_ind",
    "nbi_alta_dependencia_ind",

    # Energía
    "tiene_electricidad_ind",
    "combustible_limpio_ind",
    "combustible_solido_ind",

    # Conectividad
    "tiene_telefono_fijo_ind",
    "tiene_celular_ind",
    "tiene_tv_cable_ind",
    "tiene_internet_ind",
    "sin_tic_ind",
    "tiene_tdt_ind",

    # Gastos
    "gasto_anual_servicios_pagado_hogar",
    "gasto_anual_servicios_donado",
    "gasto_anual_servicios_autoconsumo",
    "gasto_anual_servicios_total",
    "gasto_mensual_servicios_estimado",

    # Carencias
    "carencia_materiales_vivienda_ind",
    "carencia_hacinamiento_ind",
    "carencia_agua_segura_ind",
    "carencia_saneamiento_ind",
    "carencia_electricidad_ind",
    "carencia_combustible_limpio_ind",
    "carencia_internet_ind",
    "total_carencias_bienestar_habitacional",
    "brecha_multidimensional_ind"
]

# Solo tomar columnas existentes, por seguridad
cols_metricas_fact = [col for col in cols_metricas_fact if col in stg.columns]

fact_bienestar_habitacional[cols_metricas_fact] = stg[cols_metricas_fact]

fact_bienestar_habitacional.head()
```

### 12. Validaciones del modelo estrella

```python
print("Dim tiempo:", dim_tiempo.shape)
print("Dim ubicación:", dim_ubicacion.shape)
print("Dim vivienda:", dim_vivienda.shape)
print("Dim servicios básicos:", dim_servicios_basicos.shape)
print("Dim energía:", dim_energia.shape)
print("Dim conectividad:", dim_conectividad.shape)
print("Fact bienestar habitacional:", fact_bienestar_habitacional.shape)

assert fact_bienestar_habitacional["hogar_id"].duplicated().sum() == 0, "Hay hogares duplicados en fact"
assert len(fact_bienestar_habitacional) == len(stg), "La fact no tiene el mismo número de filas que staging"

keys = [
    "tiempo_key",
    "ubicacion_key",
    "vivienda_key",
    "servicios_basicos_key",
    "energia_key",
    "conectividad_key"
]

for key in keys:
    assert fact_bienestar_habitacional[key].notna().all(), f"Hay nulos en {key}"

print("Modelo estrella validado correctamente.")
```
