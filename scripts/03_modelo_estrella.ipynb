### 1. Cargar Staging

```python
from pathlib import Path
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", 120)
pd.set_option("display.max_rows", 120)

PATH_STG = Path.cwd().parent / "data" / "processed" / "stg_enaho_hogar_100_limpio.parquet"

stg = pd.read_parquet(PATH_STG)

print("Filas staging:", len(stg))
print("Columnas staging:", len(stg.columns))

stg.head()
```

### 2. Validaciones iniciales

```python
assert "hogar_id" in stg.columns, "No existe hogar_id"
assert stg["hogar_id"].notna().all(), "Hay hogar_id nulos"
assert stg["hogar_id"].duplicated().sum() == 0, "Hay hogar_id duplicados"

print("Staging validado correctamente.")
print("Hogares:", len(stg))
```
